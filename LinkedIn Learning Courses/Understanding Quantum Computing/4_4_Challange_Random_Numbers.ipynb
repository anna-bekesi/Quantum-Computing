```python
from qiskit import *
from qiskit.visualization import plot_bloch_multivector, array_to_latex, plot_histogram
from qiskit_aer import AerSimulator
from qiskit import QuantumCircuit, transpile
from math import pi
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#Create a quantum circuit with 3 qbits
circuit= QuantumCircuit(3)

circuit.x(0)
circuit.h(1)
circuit.rx(pi/3,2)

circuit.measure_all()

visual=circuit.draw(output='mpl')
```

```python
simulator = AerSimulator()
result = simulator.run(circuit, shots = 100000).result()
hist = plot_histogram(result.get_counts())
```
